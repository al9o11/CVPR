```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
print('\033[1m' + 'Step Activation Function' + '\033[0m')
y = np.zeros(len(x))
y[x>=0] = 1

plot_graph(y, "Step(x)")
```

```python
print('\033[1m' + 'Sigmoid Activation Function' + '\033[0m')
y = 1/(1 + np.exp(-x))

plot_graph(y, "Sigmoid(x)")
```

```python
print('\033[1m' + 'TanH Activation Fuction' + '\033[0m')
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
print('\033[1m' + 'ReLU Activation Function' + '\033[0m')
y = np.maximum(0, x)

plot_graph(y, "ReLU(x)")
```

```python
print('\033[1m' + 'ELU Activation Function' + '\033[0m')
alpha = 1.0

y = np.where(x>=-2, x, alpha*(np.exp(x)-1))

plot_graph(y, "ELU(x)")
```

```python
 def selu(x, scale=1.0507, alpha=1.6732):
    return np.where(x >= 0, scale*x, scale*(alpha*np.exp(x)-alpha))

 plot_graph(selu(x), "SELU(x)")
```
